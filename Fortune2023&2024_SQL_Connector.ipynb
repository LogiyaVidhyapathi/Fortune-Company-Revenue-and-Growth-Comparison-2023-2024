### Formatting Numerical Values

```python

def format_float(value):
    if pd.isna(value):
        return value
    roundOffVal = round(value, 0)
    return int(roundOffVal) if roundOffVal.is_integer() else roundOffVal


columns_to_format = ["2023_MarketCap_March31_M", "2023_Revenues_M", "2023_RevenuePercentChange", "2023_Profits_M", "2023_ProfitsPercentChange", 
                     "2023_Assets_M", "2024_MarketCap_March28_M", "2024_Revenues_M", "2024_RevenuePercentChange", 
                     "2024_Profits_M", "2024_ProfitsPercentChange", "2024_Assets_M"]

for columns in columns_to_format:
    fortuneDataframe[columns] = fortuneDataframe[columns].apply(format_float)

fortuneDataframe[["2023_MarketCap_March31_M", "2023_Revenues_M", "2023_RevenuePercentChange", "2023_Profits_M", "2023_ProfitsPercentChange", 
                     "2023_Assets_M", "2024_MarketCap_March28_M", "2024_Revenues_M", "2024_RevenuePercentChange", 
                     "2024_Profits_M", "2024_ProfitsPercentChange", "2024_Assets_M"]] = fortuneDataframe[["2023_MarketCap_March31_M", "2023_Revenues_M", "2023_RevenuePercentChange", "2023_Profits_M", "2023_ProfitsPercentChange", 
                     "2023_Assets_M", "2024_MarketCap_March28_M", "2024_Revenues_M", "2024_RevenuePercentChange", 
                     "2024_Profits_M", "2024_ProfitsPercentChange", "2024_Assets_M"]].fillna(0).astype('int')
```

### Statistical Analysis

### Paired T Test

```python
profit_2023 = fortuneDataframe['2023_Profits_M']
profit_2024 = fortuneDataframe['2024_Profits_M']

t_statistic, p_value = stats.ttest_rel(profit_2023, profit_2024)

print(f"t_statistic: {round(t_statistic, 2)}")
print(f"p_value: {round(p_value, 2)}")
```

- __Null Hypothesis (H₀):__ There is no significant difference between 2023_Profits and 2024_Profits.
- __Alternate Hypothesis (H₁):__ There is a significant difference between 2023_Profits and 2024_Profits.

Therefore, we __reject the Null Hypothesis (0.36 > 0.05)__ (i.e. we fail to reject the null hypothesis)
